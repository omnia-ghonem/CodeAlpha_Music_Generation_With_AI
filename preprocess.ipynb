```python
"""
preprocess.py

Purpose:
1. Read all MIDI files.
2. Extract notes and chords.
3. Convert them into sequences.
4. Save processed data for training.
"""

import os
import pickle
import numpy as np
from music21 import converter, instrument, note, chord

# ===========================
# Settings
# ===========================

DATASET_PATH = "/kaggle/input/datasets/soumikrakshit/classical-music-midi"   # Change if your dataset name is different
SEQUENCE_LENGTH = 100

# Create output folder
os.makedirs("/kaggle/working/data", exist_ok=True)

notes = []

# ===========================
# Read MIDI Files
# ===========================

print("Reading MIDI files...\n")

for root, _, files in os.walk(DATASET_PATH):

    for file in files:

        if file.endswith((".mid", ".midi")):

            file_path = os.path.join(root, file)
            print(file_path)

            try:
                midi = converter.parse(file_path)

                # Read first instrument if available
                try:
                    parts = instrument.partitionByInstrument(midi)
                    elements = parts.parts[0].recurse()
                except:
                    elements = midi.flat.notes

                # Extract notes and chords
                for element in elements:

                    if isinstance(element, note.Note):
                        notes.append(str(element.pitch))

                    elif isinstance(element, chord.Chord):
                        notes.append(
                            ".".join(str(n) for n in element.normalOrder)
                        )

            except Exception as e:
                print(f"Skipped {file}: {e}")

# ===========================
# Create Dictionary
# ===========================

pitchnames = sorted(set(notes))

note_to_int = {
    note: number
    for number, note in enumerate(pitchnames)
}

network_input = []
network_output = []

# ===========================
# Create Sequences
# ===========================

for i in range(len(notes) - SEQUENCE_LENGTH):

    sequence_in = notes[i:i + SEQUENCE_LENGTH]
    sequence_out = notes[i + SEQUENCE_LENGTH]

    network_input.append(
        [note_to_int[n] for n in sequence_in]
    )

    network_output.append(
        note_to_int[sequence_out]
    )

# ===========================
# Normalize Input
# ===========================

network_input = np.reshape(
    network_input,
    (
        len(network_input),
        SEQUENCE_LENGTH,
        1
    )
)

network_input = network_input / float(len(pitchnames))

# ===========================
# Save Data
# ===========================

pickle.dump(
    notes,
    open("/kaggle/working/data/notes.pkl", "wb")
)

pickle.dump(
    pitchnames,
    open("/kaggle/working/data/pitchnames.pkl", "wb")
)

pickle.dump(
    network_input,
    open("/kaggle/working/data/network_input.pkl", "wb")
)

pickle.dump(
    network_output,
    open("/kaggle/working/data/network_output.pkl", "wb")
)

# ===========================
# Finished
# ===========================

print("\n========== DONE ==========")
print("Total notes:", len(notes))
print("Unique notes:", len(pitchnames))
print("Training sequences:", len(network_input))

print("\nSaved files:")

print("notes.pkl")
print("pitchnames.pkl")
print("network_input.pkl")
print("network_output.pkl")
```
