```python
"""
train.py

Load processed data
Train LSTM model
Save trained model
"""

import os
import pickle
import sys
sys.path.append("/kaggle/input/models/omnia00nabil/model/tensorflow2/default/1/")
from model import build_model
import numpy as np




# ==========================
# Load Data
# ==========================

DATA_PATH = "/kaggle/input/datasets/omnia00nabil/"




X = pickle.load(
    open(os.path.join(DATA_PATH,
    "network-input/network_input.pkl"), "rb")
)
y = pickle.load(
    open(os.path.join(DATA_PATH,
    "network-output/network_output.pkl"), "rb")
)
X = np.asarray(X, dtype=np.float32)
y = np.asarray(y, dtype=np.int32)
pitchnames = pickle.load(
    open(os.path.join(DATA_PATH,
    "pitchnames/pitchnames.pkl"), "rb")
)

print("Training samples :", len(X))
print("Unique notes :", len(pitchnames))

# ==========================
# Build Model
# ==========================

model = build_model(
    X.shape[1:],
    len(pitchnames)
)

model.summary()

# ==========================
# Train
# ==========================
import numpy as np

print("===== X =====")
print(type(X))
print(X.dtype)
print(X.shape)

print("\n===== y =====")
print(type(y))
print(type(y[0]))
print(len(y))

print("\n===== TensorFlow =====")
import tensorflow as tf
print(tf.__version__)
```

```python

# ==========================
# Train
# ==========================

history = model.fit(
    X,
    y,
    epochs=100,
    batch_size=32
)

# ==========================
# Save Model
# ==========================

MODEL_PATH = "/kaggle/working/models"

os.makedirs(MODEL_PATH, exist_ok=True)

model.save(
    os.path.join(
        MODEL_PATH,
        "music_model.keras"
    )
)

print("\nTraining Finished")
print("Model Saved Successfully")
```
